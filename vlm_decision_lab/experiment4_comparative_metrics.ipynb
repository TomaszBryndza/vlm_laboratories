# Experiment 4: Comparative Metrics Aggregation

Aggregate outputs from previous experiments (assumes they have been run in the same environment) and compute summary tables.

```python
import os, json, glob
from utils import parse_action_json, flip_rate, longest_streak
OUTPUT_DIR = 'outputs'  # if you saved earlier outputs; adapt path
files = glob.glob(os.path.join(OUTPUT_DIR,'*.txt'))
files[:5]
```

```python
# Example placeholder aggregation (adjust once outputs are saved)
action_sets = []
for f in files:
    txt = open(f,'r',encoding='utf-8').read()
    pj = parse_action_json(txt)
    if pj: action_sets.append(pj['action'])
print({'count_parsed': len(action_sets)})
```

```python
# Sequence metrics example
fr = flip_rate(action_sets)
st = longest_streak(action_sets)
print({'flip_rate': fr, 'longest_streak': st})
```

# Experiment 6: Sequence Consistency

Analyze stability of decisions across consecutive frames.

```python
MODEL_KEY = 'qwen25'
PROMPT = 'You are a driving assistant. What should the robot do in this situation? Output ONLY JSON {"action": "FORWARD|LEFT|RIGHT|STOP|SLOW", "rationale": "<short reason>"}. Use visible evidence only.'
```

```python
from PIL import Image
from utils import list_frame_pairs, generate_action, parse_action_json, flip_rate, longest_streak
pairs = list_frame_pairs('example_sequence_samples')
cam_paths = [p[0] for p in pairs]
cam_paths
```

```python
actions = []
parsed = []
for cam_path in cam_paths:
    img = Image.open(cam_path).convert('RGB')
    out = generate_action(MODEL_KEY, [img], PROMPT)
    pj = parse_action_json(out) or {}
    actions.append(pj.get('action','?'))
    parsed.append(pj)
actions
```

```python
fr = flip_rate(actions)
st = longest_streak(actions)
print({'flip_rate': fr, 'longest_streak': st})
```

```python
# Persist sequence metrics
from utils import save_text
if 'actions' in globals():
    import json, os
    summary = {'actions': actions, 'flip_rate': fr, 'longest_streak': st}
    save_text(os.path.join(OUTPUT_DIR,'experiment6_sequence_consistency_summary.json'), json.dumps(summary, indent=2))
    summary
else:
    print('Run previous cells first.')
```
